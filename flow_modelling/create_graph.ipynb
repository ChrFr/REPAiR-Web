```python
import networkx as nx
```

```python
G = nx.DiGraph()
```

```python
G.add_node('Import', inflow=0, outflow=50)
G.add_edge('Import', 'Selling', flow=50)
G.add_node('Production', inflow=0, outflow=130)
G.add_edge('Production', 'Export', flow=20)
G.add_edge('Production', 'Waste Management', flow=10)
G.add_edge('Production', 'Consuming', flow=10)
G.add_edge('Production', 'Selling and Consuming', flow=10)
G.add_edge('Production', 'Packaging', flow=70)
G.add_edge('Production', 'Stock 1', flow=10)
G.add_node('Stock 1', inflow=10, outflow=0)
G.add_node('Consuming', inflow=60, outflow=55)
G.add_edge('Consuming', 'Return Logistics', flow=20)
G.add_edge('Consuming', 'Collection', flow=30)
G.add_edge('Consuming', 'Export', flow=5)
G.add_node('Return Logistics', inflow=30, outflow=30)
G.add_edge('Return Logistics', 'Production of packaging', flow=30)
G.add_node('Production of packaging', inflow=30, outflow=55)
G.add_edge('Production of packaging', 'Packaging', flow=50)
G.add_edge('Production of packaging', 'Waste Management', flow=5)
G.add_node('Packaging', inflow=120, outflow=140)
G.add_edge('Packaging', 'Distribution', flow=140)
G.add_node('Distribution', inflow=140, outflow=100)
G.add_edge('Distribution', 'Selling and Consuming', flow=50)
G.add_edge('Distribution', 'Export', flow=10)
G.add_edge('Distribution', 'Selling', flow=30)
G.add_edge('Distribution', 'Stock 2', flow=10)
G.add_node('Stock 2', inflow=10, outflow=0)
G.add_node('Selling and Consuming', inflow=60, outflow=55)
G.add_edge('Selling and Consuming', 'Stock 4', flow=5)
G.add_edge('Selling and Consuming', 'Return Logistics', flow=10)
G.add_edge('Selling and Consuming', 'Collection', flow=30)
G.add_edge('Selling and Consuming', 'Waste Management', flow=10)
G.add_node('Stock 4', inflow=5, outflow=0)
G.add_node('Selling', inflow=80, outflow=80)
G.add_edge('Selling', 'Waste Management', flow=10)
G.add_edge('Selling', 'Export', flow=10)
G.add_edge('Selling', 'Consuming', flow=50)
G.add_edge('Selling', 'Stock 3', flow=10)
G.add_node('Stock 3', inflow=10, outflow=0)
G.add_node('Collection', inflow=60, outflow=60)
G.add_edge('Collection', 'Waste Management', flow=50)
G.add_edge('Collection', 'Stock 5', flow=10)
G.add_node('Stock 5', inflow=10, outflow=0)
G.add_node('Export', inflow=45, outflow=0)
G.add_node('Waste Management', inflow=85, outflow=0)

```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (15, 11)
pos = nx.kamada_kawai_layout(G)
nx.draw_networkx(G, pos)
```
